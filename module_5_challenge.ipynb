```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020_df = pd.read_csv("Resources/athletic_sales_2020.csv")
sales_2021_df = pd.read_csv("Resources/athletic_sales_2021.csv")
```

```python
# Display the 2020 sales DataFrame
sales_2020_df
```

```python
# Display the 2021 sales DataFrame
sales_2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020_df["total_sales"].dtype
```

```python
# Check the 2021 sales data types.
sales_2021_df["total_sales"].dtype
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales_df = pd.concat([sales_2020_df, sales_2021_df], axis = "rows", join = "inner")

combined_sales_reset_df = combined_sales_df.reset_index()

combined_sales_reset_df
```

```python
# Check if any values are null.
null_df = combined_sales_reset_df.isnull().sum()

null_df
```

```python
# Check the data type of each column
combined_sales_reset_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales_reset_df["invoice_date"] = pd.to_datetime(combined_sales_reset_df["invoice_date"], errors='coerce' )

```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_sales_reset_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
grouped_sales_df = combined_sales_reset_df.groupby(["region", "state", "city"])["units_sold"].sum().reset_index(name="Total_Products_Sold")

grouped_sales_sorted_df = grouped_sales_df.sort_values(by="Total_Products_Sold" , ascending=False)

# Show the top 5 results.
grouped_sales_sorted_df.head()                        
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_grouped_sales_df = combined_sales_reset_df.pivot_table(
                                                values = "units_sold",
                                                index = ["region", "state", "city"],
                                                aggfunc = "sum").rename(columns = {"units_sold": "Total_Products_Sold"})

# Rename the "units_sold" column to "Total_Products_Sold"
# See above.

# Show the top 5 results.
pivot_grouped_sales_sorted_df = pivot_grouped_sales_df.sort_values(by = "Total_Products_Sold", ascending=False)

pivot_grouped_sales_sorted_df.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_sales_region_df = combined_sales_reset_df.groupby(["region", "state", "city"])["total_sales"].sum().reset_index()

grouped_sales_region_df = grouped_sales_region_df.rename(columns={"total_sales": "Total_Sales"})

grouped_sales_region_sorted_df = grouped_sales_region_df.sort_values(by="Total_Sales", ascending=False)
# Show the top 5 results.
grouped_sales_region_sorted_df.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_grouped_region_df = combined_sales_reset_df.pivot_table(
                                                    values = "total_sales",
                                                    index = ["region", "state", "city"],
                                                    aggfunc = "sum").rename(columns = {"total_sales": "Total_Sales"})

# Optional: Rename the "total_sales" column to "Total Sales"
# See above.

# Show the top 5 results.
pivot_grouped_region_sorted_df = pivot_grouped_region_df.sort_values(by = "Total_Sales", ascending=False)

pivot_grouped_region_sorted_df.head()
# Chart has correct info, but does not look the same as the others. Consult TA before submitting.
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_retail_df = combined_sales_reset_df.groupby(["retailer", "region", "state", "city"])["total_sales"].sum().reset_index()

grouped_retail_df = grouped_retail_df.rename(columns={"total_sales": "Total_Sales"})

grouped_retail_sorted_df = grouped_retail_df.sort_values(by="Total_Sales", ascending=False)
# Show the top 5 results.
grouped_retail_sorted_df.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
pivot_grouped_retail_df = combined_sales_reset_df.pivot_table(
                                                    values = "total_sales",
                                                    index = ["retailer", "region", "state", "city"],
                                                    aggfunc = "sum").rename(columns = {"total_sales": "Total_Sales"})

# Optional: Rename the "total_sales" column to "Total Sales"
# See above.

# Show the top 5 results.
pivot_grouped_retail_sorted_df = pivot_grouped_retail_df.sort_values(by = "Total_Sales", ascending=False)

pivot_grouped_retail_sorted_df.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_athletic_df = combined_sales_reset_df[combined_sales_reset_df["product"].str.contains("Women's Athletic Footwear")]
#df.loc[(df["column_name"] =="<value>")]
womens_athletic_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_athletic_grouped_df = womens_athletic_df.groupby(["retailer", "region", "state", "city"])["units_sold"].sum().reset_index()

womens_athletic_grouped_df = womens_athletic_grouped_df.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})

womens_athletic_grouped_sorted_df = womens_athletic_grouped_df.sort_values(by="Womens_Footwear_Units_Sold", ascending=False)

# Show the top 5 results.
womens_athletic_grouped_sorted_df.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_womens_grouped_df = womens_athletic_df.pivot_table(
                                                    values = "units_sold",
                                                    index = ["retailer", "region", "state", "city"],
                                                    aggfunc = "sum").rename(columns = {"units_sold": "Womens_Footwear_Units_Sold"})

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
# See above.
# Show the top 5 results.
pivot_womens_grouped_sorted_df = pivot_womens_grouped_df.sort_values(by = "Womens_Footwear_Units_Sold", ascending=False)

pivot_womens_grouped_sorted_df.head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivot_womens_grouped_day_df = womens_athletic_df.pivot_table(
                                                    values = "total_sales",
                                                    index = ["invoice_date"],
                                                    aggfunc = "sum").rename(columns = {"total_sales": "Total_Sales"})

# Optional: Rename the "total_sales" column to "Total Sales"
# See above.

# Show the table.
pivot_womens_grouped_day_df
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
womens_daily_sales_df = pivot_womens_grouped_day_df.resample("D").sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
womens_daily_sales_sorted_df = womens_daily_sales_df.sort_values(by="Total_Sales", ascending=False)
# Notes on Challenge Rubric state descending order, notes on notebook state ascending order.
womens_daily_sales_sorted_df.head(10)
# head was changed to reflect example on Module 5 Challenge instructions.
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
womens_weekly_sales_df = pivot_womens_grouped_day_df.resample("W-SUN").sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
womens_weekly_sales_sorted_df = womens_weekly_sales_df.sort_values(by="Total_Sales", ascending=False)

womens_weekly_sales_sorted_df.head(10)
```
